```python
#Importing a pandas and numpy
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Making an empty dataframe
data = pd.DataFrame()
print(data)
```

```python
#Using numpy randint to create random variables within dataset
mydata = pd.DataFrame(np.random.randint(80,100, size=(5,4)),
                        index = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday'],
                        columns = ['Miami', 'West Palm Beach', 'Sarasota', 'Jacksonville'])

mydata
```

```python
for x in mydata:
    print(x)
```

```python
values = mydata.get_values()
values
```

```python
print(values >= 90)
```

```python
 print(values == 95)
```

```python
#Creating a boxplot for mydata
box_plot = mydata.boxplot(column = ['Miami', 'West Palm Beach', 'Sarasota', 'Jacksonville'])
```

```python
#Creating a histogram with mydata
Histo_graph = mydata.hist(column = ['Miami', 'West Palm Beach', 'Sarasota', 'Jacksonville'])
```

```python
#Creating a plot graph with mydata
plot_graph = mydata.plot()
```

```python
#Creating a bar graph with mydata
bar_graph = mydata.plot(kind ='bar',)
```
